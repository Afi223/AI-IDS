```python
import pandas as pd

path_train = r"C:\Users\abdiq\Desktop\AI_IDS\UNSW_NB15_training-set.csv"
path_test  = r"C:\Users\abdiq\Desktop\AI_IDS\UNSW_NB15_testing-set.csv"

df_train = pd.read_csv(path_train)
df_test  = pd.read_csv(path_test)

print("Loaded successfully!")
print("Training shape:", df_train.shape)
print("Testing shape:", df_test.shape)

```

```python
df_train.head()
```

```python
# Look at basic info about columns, types, and missing values
df_train.info()

print("\nMissing values per column:")
print(df_train.isna().sum())

# Show first few rows to understand structure
df_train.head()
```

```python
# Drop IDs, IP addresses, and attack category text labels (we’ll use numeric label column instead)
drop_cols = ['id', 'attack_cat', 'srcip', 'dstip']
df_train = df_train.drop(columns=drop_cols, errors='ignore')
df_test  = df_test.drop(columns=drop_cols, errors='ignore')

print("✅ Columns after dropping:", len(df_train.columns))
```

```python
from sklearn.preprocessing import LabelEncoder

# Combine to encode consistently
combined = pd.concat([df_train, df_test], axis=0)

# Encode all categorical columns at once
for col in combined.select_dtypes(include='object').columns:
    encoder = LabelEncoder()
    combined[col] = encoder.fit_transform(combined[col].astype(str))

# Split them back into train/test
df_train = combined.iloc[:len(df_train), :].copy()
df_test  = combined.iloc[len(df_train):, :].copy()

print("Encoded categorical columns safely for both train and test.")

```

```python
X_train = df_train.drop('label', axis=1)
y_train = df_train['label']

X_test = df_test.drop('label', axis=1)
y_test = df_test['label']
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

iso = IsolationForest(
    contamination=0.1,  # expected proportion of attacks
    random_state=42,
    n_estimators=200
)
iso.fit(X_train)

# Predict: -1 = anomaly -> 1 (attack), 1 = normal -> 0
y_pred = [1 if x == -1 else 0 for x in iso.predict(X_test)]

print(classification_report(y_test, y_pred, target_names=['Normal','Attack']))
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix – Isolation Forest IDS')
plt.show()
```

```python
from sklearn.metrics import roc_auc_score, roc_curve

# Isolation Forest scores (lower = more anomalous)
scores = -iso.score_samples(X_test)

fpr, tpr, thresholds = roc_curve(y_test, scores)
roc_auc = roc_auc_score(y_test, scores)

plt.figure(figsize=(7,6))
plt.plot(fpr, tpr, color='navy', label=f'ROC curve (AUC = {roc_auc:.3f})')
plt.plot([0,1],[0,1],'--',color='gray')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve – Isolation Forest IDS')
plt.legend(loc='lower right')
plt.grid(alpha=0.3)
plt.show()

print(f"ROC-AUC Score: {roc_auc:.3f}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=15,
    random_state=42,
    n_jobs=-1
)
rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)

print("=== Random Forest Results ===")
print(classification_report(y_test, y_pred_rf, target_names=['Normal','Attack']))

cm_rf = confusion_matrix(y_test, y_pred_rf)
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Greens')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix – Random Forest IDS')
plt.show()
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score

# Recalculate accuracy and ROC for both models
acc_iso = accuracy_score(y_test, y_pred)
acc_rf  = accuracy_score(y_test, y_pred_rf)

roc_iso = roc_auc_score(y_test, -iso.score_samples(X_test))
roc_rf  = roc_auc_score(y_test, rf.predict_proba(X_test)[:,1])

results = pd.DataFrame({
    'Model': ['Isolation Forest', 'Random Forest'],
    'Accuracy': [acc_iso, acc_rf],
    'ROC-AUC': [roc_iso, roc_rf]
})
display(results)

# Optional bar chart
results.set_index('Model').plot(kind='bar', figsize=(6,4), color=['#1f77b4','#2ca02c'])
plt.title('Model Performance Comparison')
plt.ylabel('Score')
plt.ylim(0,1)
plt.grid(axis='y', alpha=0.3)
plt.show()
```
